```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("carprices.csv")
dataset.head(5)
```

```python
dataset.info() , dataset.columns
```

```python
dummie = pd.get_dummies(dataset["Car Model"])
dummie
```

```python
dummie = dummie.replace({True :1 , False:2})
dummie
```

```python
merged = pd.concat([dataset,dummie] , axis="columns")
merged
```

```python
final = merged.drop("Car Model", axis="columns")
final
```

```python
x = final.drop("Mercedez Benz C class", axis="columns")
x
```

```python
x = x.drop(["Sell Price($)"], axis="columns")
x
```

```python
y = final["Sell Price($)"]
y
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(x,y)
```

```python
model.predict([[45000,4,0,1]])
```

```python
model.score(x,y)*100
```
